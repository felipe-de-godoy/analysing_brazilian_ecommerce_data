# Sellers And Buyers Analize

## Sellers Analize

```python
df = pd.merge(left = sellers, 
              right= items,
              how = 'left',
             on = 'seller_id')
df.head()
```

```python
cols = ['seller_city']
df[cols] = df[cols].apply(lambda x: x.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8'))
```

```python
df['Total_sales'] = (df['price'].fillna(0) + df['freight_value'].fillna(0))*(df['order_item_id'].fillna(0))
```

```python
df.groupby('seller_id').sum()['Total_sales']

top_sellers = pd.merge(left = pd.DataFrame(df.groupby('seller_id').sum()['Total_sales']), 
         right =df.drop('Total_sales', axis =1),
         left_index= True,
         right_on='seller_id'       
        )[['Total_sales','seller_id','seller_city','seller_state']]
top_sellers.drop_duplicates(inplace = True)
top_sellers.sort_values(['Total_sales'], ascending = False, inplace = True)
top_sellers.head(10)
```

## Buyers Analize

```python
df2 = pd.merge(right = orders,
               left = items , 
               on= 'order_id',
               how = 'left')

df3 = pd.merge(right = df2, 
               left = customers, 
               on = 'customer_id',
               how = 'left')
df3['Total_purchases'] = df3['order_item_id'].fillna(0)*(df3['price'].fillna(0)+ df3['freight_value'].fillna(0))
```

```python
cols = ['customer_city']
df3[cols] = df3[cols].apply(lambda x: x.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8'))
```

```python
df3.groupby('customer_id').sum()['Total_purchases']

top_customers = pd.merge(left = pd.DataFrame(df3.groupby('customer_id').sum()['Total_purchases']), 
         right =df3.drop(['Total_purchases'], axis = 1),
         left_index= True,
         right_on='customer_id'       
         )[['Total_purchases','customer_id','customer_city','customer_state']]
top_customers.drop_duplicates(inplace = True)
top_customers.sort_values(['Total_purchases'], ascending = False, inplace = True)
top_customers.head(10)
```

```python
geolocation = pd.read_csv('olist_geolocation_dataset.csv')

geolocation.sort_values(by = 'geolocation_zip_code_prefix')

# fixes special characters
# deletes duplicates from zip code lat/long

geolocation[['geolocation_city']] = geolocation[['geolocation_city']].apply(lambda x: x.str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8'))
geolocation = geolocation.groupby(['geolocation_zip_code_prefix', 'geolocation_city','geolocation_state']).mean() 
geolocation.sort_values(by = 'geolocation_city', inplace=True)
geolocation.reset_index(level=['geolocation_city','geolocation_state'], inplace=True)
```

```python
def map_lat_long(df,MAX_RECORDS=50 ):
    
    m = folium.Map(
        location=[-16, -60],    # Centro do Brasil    
        zoom_start=4)
    
    for idx in range(min(MAX_RECORDS, df.shape[0])):
        folium.Marker([df[['geolocation_lat']].iloc[idx][0],
            df[['geolocation_lng']].iloc[idx][0]]).add_to(m)
    return(m)
```

```python
def boxplot_mean_std(base, var):
    sns.boxenplot(base[var], )
    plt.xscale('log')
    plt.show()
    print("Median:             "+str(base[var].median()))
    print("Mean:               "+ str(base[var].mean()))
    print("Standar Deviation:  " + str(base[var].std()))



boxplot_mean_std(df_freight2[df_freight2.density<30],'density' )

boxplot_mean_std(df_freight2,'freight_value' )

boxplot_mean_std(df_freight2,'price' )

boxplot_mean_std(df_freight2,'volume_cm3_per_thousand' )
```

## Clustering

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot   as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
df_freight_cluster = df_freight2[['volume_cm3_per_thousand', 'product_weight_kg', 'freight_value']]
df_freight_cluster = df_freight_cluster.dropna()

modelo = KMeans(n_clusters=6,   
                n_jobs = 4,                 
                max_iter = 300, 
                random_state= 1)
df_freight_cluster['cluster'] = modelo.fit_predict(df_freight_cluster)


#ordena nome dos clusters por tamanho/peso
y = df_freight_cluster.groupby('cluster').mean().sort_values('freight_value')
df_freight_cluster['cluster'] = df_freight_cluster['cluster']*100

for i in y.index:
    if i != 0:
        df_freight_cluster.loc[(df_freight_cluster['cluster'] == y.index[i]*100), 'cluster'] = i
df_freight_cluster.loc[(df_freight_cluster['cluster'] == y.index[0]*100), 'cluster'] = 0
df_freight_cluster.head()
```

```python
cols = ['product_weight_kg', 'volume_cm3_per_thousand','freight_value']
for col in cols:
    sns.boxenplot(y = col, x = 'cluster', data = df_freight_cluster)
    plt.show()
```

```python
fig = plt.figure()
ax = Axes3D(fig)
fig.subplots_adjust(bottom=-0.15,top=1.2)
ax.scatter(
    df_freight_cluster['volume_cm3_per_thousand'], 
    df_freight_cluster['product_weight_kg'], 
    df_freight_cluster['freight_value'],
    s = 10, c = df_freight_cluster['cluster'])
ax.set_xlabel('volume_cm3_per_thousand')
ax.set_ylabel('Weight')
ax.set_zlabel('Freight')
plt.show()
```

## Predicting Freight

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
def stats_continuous(X_train,y_train,  X_test,y_test,lr ):
    predict = lr.predict(X_test)
    mae = mean_absolute_error(y_test, predict)
    mse = mean_squared_error(y_test, predict)
    r2 = r2_score(y_test, predict)
    print("Test Metrics:")
    print("Mean Absolute Error: " + str(mae))
    print("Mean Squared Error:  " + str(mse))
    print("R squared:           " + str(r2))

    predict = lr.predict(X_train)
    mae = mean_absolute_error(y_train, predict)
    mse = mean_squared_error(y_train, predict)
    r2 = r2_score(y_train, predict)
    print('Train Metrics: ')
    print("Mean Absolute Error: " + str(mae))
    print("Mean Squared Error:  " + str(mse))
    print("R squared:           " + str(r2))
```

```python
def freight( weight,lr):
    return(lr.predict(pd.DataFrame([(weight)], columns = ['product_weight_kg']))[0])
```

```python
X = df_freight_cluster[['volume_cm3_per_thousand']]
y = df_freight_cluster.freight_value

X_train, X_test, y_train, y_test = train_test_split(X,y)

lr = LinearRegression()
lr.fit(X_train, y_train)

stats_continuous(X_train,y_train,  X_test,y_test,lr )
```

```python
X = df_freight_cluster[['product_weight_kg']]
y = df_freight_cluster.freight_value

X_train, X_test, y_train, y_test = train_test_split(X,y)

lr = LinearRegression()
lr.fit(X_train, y_train)

stats_continuous(X_train,y_train,  X_test,y_test,lr )
```
